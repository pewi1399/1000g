```r
library(ggplot2)

power_calculator = function(n){
  
    # berakna power for proportionstest
    # notera att n ar antal per grupp!
  pwr = power.prop.test(p1 = 0.20, p2 = 0.07, n = n)$power
  
  return(pwr)
}

# spara mojliga varden for n
# notera att n ar antal per grupp!
n <- 1:370

# berakna respektive powervarde for givet n
power_level <- sapply(n, power_calculator)

# klistra ihop dataset
dat <- data.frame(cbind(n, power_level))

# plot
library(ggplot2)
ggplot(dat)+
  aes(x = n, y = power_level)+
  geom_line(col = "dodgerblue")+
  scale_y_continuous(breaks = seq(0,1,0.1))+
  scale_x_continuous(breaks = seq(0,370,50))+
  xlab("n per group")+
  theme_bw()
```
